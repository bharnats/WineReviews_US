```python
import os
import pandas as pd
import numpy as np
```

```python
# read the csv files into pandas as dataframes
filepath = os.path.join("..","raw_data/wineData.csv")
wine_df = pd.read_csv(filepath,encoding="iso-8859-1", low_memory=False)
```

```python
wine_df.head()
```

```python
wine_df = wine_df[wine_df.region_1!= ""]
```

```python
wine_df.count()
```

```python
wine_df = wine_df.drop(['region_2','designation','Unnamed: 0'],axis=1)
```

```python
wine_df.count()
```

```python
wine_df.dropna()
```

```python
wine_df.count()
```

```python
# replace the empty strings with np.nan objects
wine_df['region_1'].replace('', np.nan, inplace=True)
wine_df.tail()
```

```python
# dropping rows based on the NaN 'region_1' values 
wine_df.dropna(subset=['region_1'], inplace=True)
```

```python
wine_df.count()
```

```python
# replace the empty strings with np.nan objects in price column
wine_df['price'].replace('', np.nan, inplace=True)
wine_df.tail()
```

```python
# dropping rows based on the NaN 'price' values 
wine_df.dropna(subset=['price'], inplace=True)
```

```python
wine_df.count()
```

```python
wine_df = wine_df.drop(['taster_name','taster_twitter_handle'],axis=1)
```

```python
wine_df.count()
```

```python
regions = wine_df['region_1'].unique()
len(regions)
```

```python
variety = wine_df['variety'].unique()
```

```python
len(variety)
```

```python

regions_df = pd.DataFrame({'region_1':regions,'lat':"",'lon':""})
```

```python
regions_df.head()
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim()
```

```python
for index,row in regions_df.iterrows():
    try:
        location = geolocator.geocode(row["region_1"],timeout=10)
        lat = location.latitude
        lon = location.longitude
    
        regions_df.set_value(index,"lat",lat)
        regions_df.set_value(index,"lon",lon)
    except AttributeError:
        print("oops!! missing region_1 values")
```

```python
regions_df.head()
# regions_df.to_csv('regions.csv',encoding='utf-8')
```

```python
regions_df.to_csv('regions.csv',encoding='utf-8')
```

```python
# read the csv files into pandas as dataframes
filepath = os.path.join("regions.csv")
my_regions_df = pd.read_csv(filepath,encoding='utf-8')
```

```python
my_regions_df.head()
```

```python
#drop the 'Unnames: 0' column
my_regions_df = my_regions_df.drop(['Unnamed: 0'],axis=1)
```

```python
my_regions_df.head()
```

```python
# replace the empty strings with np.nan objects in 'lat' and 'lon' columns of regions_df dataframe
my_regions_df['lat'].replace('', np.nan, inplace=True)

```

```python
my_regions_df['lon'].replace('', np.nan, inplace=True)
```

```python
my_regions_df
```

```python
# create latitude and longitude columns in wine_df
wine_df['latitude']=''
wine_df['longitude']=''
```

```python
my_regions_df.rename(columns={'lat': 'latitude','lon':'longitude'}, inplace=True)
```

```python
coord_df = pd.merge(wine_df, my_regions_df, on='region_1')
```

```python

#drop the 'latitude_x' nad 'longitude_x' column
coord_df = coord_df.drop(['latitude_x','longitude_x'],axis=1)
```

```python
coord_df.rename(columns={'latitude_y': 'latitude','longitude_y':'longitude'}, inplace=True)
```

```python
coord_df['longitude'].replace('', np.nan, inplace=True)
coord_df['latitude'].replace('', np.nan, inplace=True)
coord_df.tail()
```
