```python

regions_df = pd.DataFrame({'region_1':regions,'lat':"",'lon':""})
```

```python
regions_df.head()
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim()
```

```python
for index,row in regions_df.iterrows():
    try:
        location = geolocator.geocode(row["region_1"],timeout=10)
        lat = location.latitude
        lon = location.longitude
    
        regions_df.set_value(index,"lat",lat)
        regions_df.set_value(index,"lon",lon)
    except AttributeError:
        print("oops!! missing region_1 values")
```

```python
# convert the grouped object to dataframe
variety_df = pd.DataFrame({'count' : coord_df.groupby( [ "variety", "region_1","latitude","longitude"] ).size()}).reset_index()
variety_df.head()
```

```python
# creating the geojson data for a grape variety chosen

geojson=[]
for index,row in variety_df.iterrows():
    feature = {
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [row['latitude'],row['longitude']] 
        },
        "properties": {
            "name":row['region_1'],
            "grape":row['variety'],
            "count":row['count']
        }
    }
    geojson.append(feature)

geoJson_data = {
  "type": "FeatureCollection",
  "features": geojson
}

```

```python
import json
with open('geoJson_data.txt', 'w') as outfile:
    json.dump(geoJson_data, outfile)

    
```

```python
# Import Dependencies
import os
import numpy as np
import pandas as pd
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine,func
```

```python
#################################################
# Database Setup
#################################################
dbfile = os.path.join('wine_reviews.sqlite')
engine = create_engine(f"sqlite:///{dbfile}")
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
# Save references to each table
wine_reviews = Base.classes.reviews 
# Create our session (link) from Python to the Database
session = Session(engine)
```

```python
# Query for the number of wine reviews by state
stmt = session.query(wine_reviews).statement
df = pd.read_sql_query(stmt, session.bind)

# df.set_index('province', inplace=True)
```

```python
df.head()
```

```python
cluster_df = df.groupby(["region_1","latitude","longitude"]).agg({"variety":pd.Series.nunique}).reset_index()
```

```python
cluster_df.head()
```

```python
# create json dictionary from the above dataframe


json_data =[]
for index,row in cluster_df.iterrows():
   
    location = {
        "type": "Point",
        "coordinates": [row['latitude'],row['longitude']] ,
        "variety_count":row['variety'],
        "region":row['region_1']
    }
    json_data.append(location)
```
