```python
import os
import pandas as pd
```

```python
# read the csv files into pandas as dataframes
filepath = os.path.join("..","raw_data/wineData.csv")
wine_df = pd.read_csv(filepath,encoding="iso-8859-1", low_memory=False)
```

```python
wine_df.head()
```

```python
wine_df["latitude"]=""
wine_df["longitude"]=""
```

```python

wine_df.head()
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim()
for index,row in wine_df.iterrows():
    try:
        location = geolocator.geocode(row["region_1"])
        lat = location.latitude
        lon = location.longitude
    
        wine_df.set_value(index,"latitude",lat)
        wine_df.set_value(index,"longitude",lon)
    except AttributeError:
        print("oops!! missing region_1 values")
    
    
```

```python
wine_df.head()
```

```python
wine_df.to_csv("lat-lon.csv",index=True)
```
